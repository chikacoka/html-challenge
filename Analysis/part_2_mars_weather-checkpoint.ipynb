### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
url = 'https://static.bc-edx.com/data/web/mars_facts/temperature.html'
browser.visit(url)
html = browser.html
mars_soup = soup(html, 'html.parser')
```

```python
mars_soup
```

```python
# Find the table
table = mars_soup.find('table', class_='table')
table
```

```python
# Extract all rows of data
rows = mars_soup.find_all('tr', class_="data-row")
rows
```

```python

# Create an empty list for the rows
mars_temp_rows = []

# Extract all rows of data by looping through the scraped data to add to the created list of rows above
for row in rows:
    table_rows = row.find_all('td')
    row_data = [cols.text for cols in table_rows]
    mars_temp_rows.append(row_data)
 
mars_temp_rows
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
mars_weather_df = pd.DataFrame(mars_temp_rows,columns=['id','terrestrial_date', 'sol', 'ls', 'month','min_temp', 'pressure'])
```

```python
# Confirm DataFrame was created successfully
mars_weather_df.head()
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column

mars_weather_df.dtypes
```

```python
# Change data types for data analysis: Convert id, sol, ls, and month to integer, and min_temp & pressure to float

updated_mars_weather_df = mars_weather_df.astype({'id':int, 'sol': int, 'ls': int, 'month': int, 'min_temp': float, 'pressure':float})
```

```python
# Confirm the data types
updated_mars_weather_df.dtypes
```

```python
# Convert terrestrial_date to datetime

from datetime import datetime as dt
updated_mars_weather_df["terrestrial_date"] = pd.to_datetime(updated_mars_weather_df["terrestrial_date"])
updated_mars_weather_df.head(10)
```

```python
# Confirm type changes were successful by examining data types again

updated_mars_weather_df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# Summary statistics of updated_mars_weather_df
updated_mars_weather_df.describe()
```

```python
updated_mars_weather_df
```

```python
# 1. How many months are there on Mars?
months_count = updated_mars_weather_df["month"].unique()
print("Below is the list of Martian months which are 12 altogether: ")
print(months_count)
```

```python
# 2. How many Martian days' worth of data are there?
martian_days_data_beginning = updated_mars_weather_df["sol"].min()
martian_days_data_ending = updated_mars_weather_df["sol"].max()

martian_days_data_range = martian_days_data_ending - martian_days_data_beginning

print(f"Number of Martian days' worth of data: {martian_days_data_range}")
print("Note: This is different from the sample count of 1867 in the previously shown summary stattistics ")
```

```python
# 3. What is the average low temperature by month?

avg_low_temp_by_month = updated_mars_weather_df.groupby(["month"]).mean("min_temp")
avg_low_temp_by_month_df = avg_low_temp_by_month.drop(['id','sol', 'ls', 'pressure'], axis=1)
print(" Average Minimum Temperature By Month: ")
avg_low_temp_by_month_df
```

```python
# Plot the average temperature by month
avg_low_temp_by_month_df.plot(kind='bar')
```

```python
# Identify the coldest and hottest months in Curiosity's location
print(avg_low_temp_by_month_df.min())
print(avg_low_temp_by_month_df.max())
print("March (month 3) has the coldest minimum temperature on Mars at -83.307292:")
print("while August (month 8) has the hottest minimum temperature at -68.382979")
```

```python
# 4. Average pressure by Martian month
avg_pressure_by_martian_month = updated_mars_weather_df.groupby(["month"]).mean("pressure")
avg_pressure_by_martian_month_df = avg_pressure_by_martian_month.drop(['id', 'sol', 'ls', 'min_temp'], axis=1)
print(" Average Pressure By Martian Month: ")
avg_pressure_by_martian_month_df
```

```python
# Plot the average pressure by month
avg_pressure_by_martian_month_df.plot(kind='bar')
```

```python
# Which months have the lowest and the highest atmospheric pressure on Mars?
print(avg_pressure_by_martian_month_df.min())
print(avg_pressure_by_martian_month_df.max())
print("June (month 6) has the lowest atmospheric pressure on Mars at 745.054422:")
print("while September (month 9) has the highest atmospheric pressure at 913.30597")
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?

updated_mars_weather_df.min_temp.plot()
plt.xlabel("Terrestrial Days")
plt.ylabel("Minimum Temperature")
```

```python
# From the above minimum temperature line plots over terrestrial days, the distance between the successive peaks is approximately 1425 - 750, i.e., roughly 675 terrestrial earth days.
# The above result falls close to NASA's Science website which confirms it at 687 Earth days or one Mars year
# Source: https://mars.nasa.gov/resources/21392/mars-in-a-minute-how-long-is-a-year-on-mars/
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
